```python
import yfinance as yf
import pandas as pd
from functools import reduce


def get_stock_period(ticker):
    tick = yf.Ticker(ticker)
    return pd.DataFrame(tick.history(period='1d', interval='1m'))['Open']

stocks = [
    {'prices': get_stock_period('GOOG'), 'name': 'GOOG'},
    {'prices': get_stock_period('BKNG'), 'name': 'BKNG'},
    {'prices': get_stock_period('WTM'), 'name': 'WTM'}
]

bankroll = 10000
money_earned = 0
threshold = 2

initial_investment = sum([stock['prices'][0] for stock in stocks])

# update your holdings
for stock in stocks:
    stock['holding'] = stock['prices'][0]

bankroll -= initial_investment


class Stock:
    def __init__(self, name):
        self.prices = get_stock_period(name)
        self.name = name
        self.purchase_price = 0
        self.amount = 0
    
    def buy(self, period_index, amount):
        self.purchase_price = self.prices[period_index]
        self.amount += amount
    
    def sell(self, period_index, amount):
        current_price = self.prices[period_index]
        earned = current_price * amount
        
        # deduct
        self.amount -= amount
        if self.amount == 0:
            self.purchase_price = 0
        
        return earned            

class StockHolder:
    def __init__(self, stocks):
        self.stocks = stocks
        
    def get_stock_names(self):
        return [stock.name for stock in self.stocks]
            
    
    def select_stock(self, name):
        # TODO check that stock name exists
        names = self.get_stock_names()
        if name not in names:
            raise AttributeError(f'stock {name} not registered')
        
        i = names.index(name)
        return self.stocks[i]

        


# for i in range(1, 388):
#     for stock in stocks:
#         stock_name = stock['name']
#         current_holding = None
#         for _stock in stocks:
#             if _stock['name'] == stock_name:
#                 current_holding = _stock['holding']
#                 break
        
#         assert current_holding is not None
        
#         if (stock[i] - current_holding) > threshold:
#             # sell it
#             money_earned = stock[i] - current_holding
#             _stock['holding'] = 0

initial_investment
```
